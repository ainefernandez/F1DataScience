```python
import fastf1
import pandas as pd

# Enable the FastF1 cache
fastf1.Cache.enable_cache("/Users/ainefernandez/documents/itam/8SEMESTRE/F1 Python")

year = 2024
sessions = range(1, 15)
position = pd.DataFrame(columns=['GP', 'Driver', 'Team', 'Gap'])

# Iterate over each race session
for race in sessions:
    session = fastf1.get_session(year, race, 'R')
    session.load()
    
    gp_name = session.event['EventName']
    
    # Get unique drivers
    drivers = pd.unique(session.laps['Driver'])
    lapdata = []

    # Iterate over each driver
    for driver in drivers:
        # Pick quickest laps for each driver
        laps = session.laps.pick_driver(driver).pick_quicklaps()
        
        # Convert LapTime to seconds
        laps['LapTime'] = laps['LapTime'].dt.total_seconds()
        
        # Append lap data
        lapdata.append(laps)

    # Concatenate all lap data for the current race
    all_laps = pd.concat(lapdata, ignore_index=True)
    
    # Calculate average lap times for each driver in the current race
    average_laptimes = all_laps.groupby(['Team', 'Driver'])['LapTime'].mean().reset_index()
    
    # Find the fastest average lap time in the current race
    fastest_avg_race_pace = average_laptimes['LapTime'].min()
    
    # Calculate the gap to the fastest average race pace for each driver in the current race
    average_laptimes['GapToFastest'] = average_laptimes['LapTime'] - fastest_avg_race_pace
    
    # Add GP name to the DataFrame
    average_laptimes['GP'] = gp_name
    
    # Append results to the main DataFrame
    position = pd.concat([position, average_laptimes[['GP', 'Driver', 'Team', 'GapToFastest']]], ignore_index=True)

# Sort by team and GP
position = position.sort_values(by=['Team', 'GP'])

print(position)

# Optionally save to CSV
# position.to_csv("GapToFastestRacePacePerGP_2024.csv", index=False)
```

```python
from itertools import product

# Configuracin inicial
teams = {
    "McLaren": 608,
    "Ferrari": 584,
    "Red Bull": 555
}

# Puntos para sprint y carrera
sprint_points = [8, 7, 6, 5, 4, 3, 2, 1]  # Del 1 al 8 en sprint
race_points = [25, 18, 15, 12, 10, 8, 6, 4, 2, 1]  # Del 1 al 10 en carrera

# Generar combinaciones de puntos posibles para las dos carreras
# Qatar: sprint y carrera | Abu Dhabi: solo carrera
qatar_scenarios = product(sprint_points + [0], race_points)
abu_dhabi_scenarios = race_points

# Calcular los escenarios combinados
all_scenarios = list(product(qatar_scenarios, abu_dhabi_scenarios))

# Evaluar cada escenario
winning_scenarios = {
    "McLaren": [],
    "Ferrari": [],
    "Red Bull": []
}

for scenario in all_scenarios:
    (qatar_sprint, qatar_race), abu_dhabi_race = scenario
    
    # Sumar puntos para cada equipo
    mclaren_points = teams["McLaren"] + qatar_sprint + qatar_race + abu_dhabi_race
    ferrari_points = teams["Ferrari"] + qatar_sprint + qatar_race + abu_dhabi_race
    red_bull_points = teams["Red Bull"] + qatar_sprint + qatar_race + abu_dhabi_race
    
    # Determinar ganador
    max_points = max(mclaren_points, ferrari_points, red_bull_points)
    if mclaren_points == max_points and mclaren_points > ferrari_points and mclaren_points > red_bull_points:
        winning_scenarios["McLaren"].append((qatar_sprint, qatar_race, abu_dhabi_race))
    elif ferrari_points == max_points and ferrari_points > mclaren_points and ferrari_points > red_bull_points:
        winning_scenarios["Ferrari"].append((qatar_sprint, qatar_race, abu_dhabi_race))
    elif red_bull_points == max_points and red_bull_points > mclaren_points and red_bull_points > ferrari_points:
        winning_scenarios["Red Bull"].append((qatar_sprint, qatar_race, abu_dhabi_race))

# Mostrar resultados
for team, scenarios in winning_scenarios.items():
    print(f"Escenarios para que {team} gane el campeonato:")
    for scenario in scenarios[:5]:  # Mostrar solo los primeros 5 escenarios por equipo
        qatar_sprint, qatar_race, abu_dhabi_race = scenario
        print(f"  Qatar Sprint: {qatar_sprint}, Qatar Carrera: {qatar_race}, Abu Dhabi Carrera: {abu_dhabi_race}")
    print(f"Total de escenarios: {len(scenarios)}")
    print()

    
```
