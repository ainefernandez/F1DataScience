```python
#Avg gap to pole 
fastf1.Cache.enable_cache("/Users/ainefernandez/documents/itam/8SEMESTRE/F1 Python")

year = 2024
ses = 'Q'
position = pd.DataFrame(columns=['Driver', 'Gap'])

for i in range(1,15):
    session = fastf1.get_session(year, i, ses)
    session.load()
    drivers = pd.unique(session.laps['Driver'])
    list_fastest_laps = list()
    
    for drv in drivers:
        drvs_fastest_lap = session.laps.pick_driver(drv).pick_fastest()
        list_fastest_laps.append(drvs_fastest_lap)
    
    fastest_laps = Laps(list_fastest_laps).sort_values(by='LapTime').reset_index(drop=True)
    pole_lap = fastest_laps.iloc[0]  # Get the pole position lap
    
    # Calculate gaps and store in the position DataFrame
    for index, row in fastest_laps.iterrows():
        gap = (row['LapTime'] - pole_lap['LapTime']).total_seconds()  # Calculate gap in seconds
        pos = {'Driver': row['Driver'], 'Gap': gap}
        position = position.append(pos, ignore_index=True)

print(position)
```

```python
import fastf1
import pandas as pd

# Enable cache
fastf1.Cache.enable_cache("/Users/ainefernandez/documents/itam/8SEMESTRE/F1 Python")

# Set parameters
year = 2024
ses = 'R'

# DataFrame to hold positions
position = pd.DataFrame(columns=['Driver', 'Laps', 'Team'])

# Loop through races
for i in range(1, 15):  # Assuming there are up to 14 races
    try:
        session = fastf1.get_session(year, i, ses)
        session.load()
        drivers = pd.unique(session.results['Abbreviation'])
        for driver in drivers:
            laps = session.laps.pick_driver(driver)
            laps_led = laps[laps["Position"] == 1].shape[0]
            team = laps.iloc[0]["Team"] if not laps.empty else "Unknown"
            pos = {'Driver': driver, 'Laps': laps_led, 'Team': team}
            position = position.append(pos, ignore_index=True)
    except Exception as e:
        print(f"Skipping session {i} due to an error: {e}")

# Group by driver and sum the laps led
grouped_position = position.groupby("Driver").agg({"Laps": "sum"}).reset_index()

# Calculate the total laps led across all drivers
total_laps_led = grouped_position["Laps"].sum()

# Calculate the percentage of laps led by each driver
grouped_position["Percentage"] = (grouped_position["Laps"] / total_laps_led) * 100

print(grouped_position)

# Optionally save to CSV
# grouped_position.to_csv("LapsLedPercentage_ByDriver.csv", index=False)
```

```python
import fastf1
import pandas as pd

fastf1.Cache.enable_cache("/Users/ainefernandez/documents/itam/8SEMESTRE/F1 Python")

year = 2023
ses = 'R'
position = pd.DataFrame(columns=['Driver', 'Gap'])

exclude_race_number = 21  # Change this to the race number you want to exclude

for i in range(13, 23):
    if i == exclude_race_number:
         continue  # Skip the excluded race number

    session = fastf1.get_session(year, i, ses)
    session.load()

    drivers = pd.unique(session.laps['Driver'])
    list_quick_laps = list()

    for drv in drivers:
        drvs_laps = session.laps.pick_driver(drv).pick_quicklaps()
        avg = drvs_laps['LapTime'].mean()  # Use mean() for average lap time
        list_quick_laps.append({'Driver': drv, 'AvgLapTime': avg})

    avg_laps_df = pd.DataFrame(list_quick_laps)
    avg_laps_df = avg_laps_df.sort_values(by='AvgLapTime').reset_index(drop=True)
    
    # Get the pole position driver
    pole_driver = avg_laps_df.iloc[0]['Driver']

    # Calculate gaps and store in the position DataFrame
    avg_laps_df['Gap'] = avg_laps_df['AvgLapTime'] - avg_laps_df.loc[avg_laps_df['Driver'] == pole_driver, 'AvgLapTime'].values[0]
    avg_laps_df['Gap'] = avg_laps_df['Gap'].dt.total_seconds()
    
    position = position.append(avg_laps_df[['Driver', 'Gap']], ignore_index=True)

print(position)
```

```python
grouped_position = position.groupby('Driver')
grouped_position_mean = grouped_position['Gap'].mean()
print(grouped_position_mean)

grouped_position_mean.to_csv("SPSAVGGapRacePace.csv")
```
