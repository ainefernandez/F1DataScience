```python
import fastf1
import pandas as pd

fastf1.Cache.enable_cache("/Users/ainefernandez/documents/itam/8SEMESTRE/F1 Python")

year = 2023
wknd = "Saudi Arabia"
ses = "Q"

session = fastf1.get_session(year, wknd, ses)
session.load()

drivers = pd.unique(session.laps['Driver'])

avgspeed = pd.DataFrame(columns=['X', 'Y', 'Speed'])

for driver in drivers:
    lap = session.laps.pick_driver(driver).pick_fastest()

    # Telemetry
    x = lap.telemetry['X']
    y = lap.telemetry['Y']
    speed = lap.telemetry['Speed']

    tel = pd.DataFrame({'X': x, 'Y': y, 'Speed': speed})
    
    # Calculate average speed for each X, Y position
    avg_speed = tel.groupby(['X', 'Y'])['Speed'].mean().reset_index()
    
    # Append to the overall dataframe
    avgspeed = avgspeed.append(avg_speed, ignore_index=True)

# Now avgspeed DataFrame contains the average speed for each X, Y position across all drivers
print(avgspeed)

```

```python
import fastf1
import pandas as pd

fastf1.Cache.enable_cache("/Users/ainefernandez/documents/itam/8SEMESTRE/F1 Python")

year = 2023
position = pd.DataFrame(columns=['Driver', 'Points', 'Year', 'Podium'])
drivers = ["VER", "PER", "HAM", "RUS", "SAI",
           "LEC", "NOR", "PIA", "ALO", "STR",
           "GAS", "OCO", "ALB", "SAR", "RIC",
           "TSU", "BOT", "ZHO", "MAG", "HUL"]
years=[2019,2020,2021,2022,2023]


for year in years:
    # Race session
    race_session = fastf1.get_session(year, "Abu Dhabi", 'R')
    race_session.load()

    for driver in drivers:
        # Check if driver participated in the race
        if driver not in race_session.laps['Driver'].unique():
            continue

        results = race_session.get_driver(driver)
        race_points = results['Points']

        # Check if 'Position' is numeric; convert if needed
        race_pos = pd.to_numeric(results['Position'], errors='coerce')

        # Check if the race is a podium finish
        podium = 1 if race_pos in [1, 2, 3] else 0

        # Create DataFrame row
        pos = {'Driver': driver, 'Points': race_points, 'Year': year, 'Podium': podium}
        position = position.append(pos, ignore_index=True)


```

```python
grouped_position = position.groupby('Driver')
grouped_position_mean = grouped_position['Points'].mean()
grouped_position_sum = grouped_position['Points'].sum()
podiums=grouped_position['Podium'].sum()
print("Top 5 Drivers by Sum of Points:")
print(grouped_position_sum)
print("Mean Points:")
print(grouped_position_mean)

print("Podiums:")
print(podiums)


top_5_drivers_mean = grouped_position_mean.nlargest(5).reset_index()
print("Top 5 Drivers by Mean Points:")
print(top_5_drivers_mean)

#top_5_drivers_sum = grouped_position_sum.nlargest(5).reset_index()

#print(top_5_drivers_sum)

position_2020 = position[position['Year'] == 2020]
print("Data for the Year 2020:")
print(position_2020)

#grouped_position_mean.to_csv("FSAVGRace.csv")
```

```python
import fastf1

import matplotlib.pyplot as plt
import numpy as np


session = fastf1.get_session(2023, 'Bahrain', 'Q')
session.load()

lap = session.laps.pick_fastest()
pos = lap.get_pos_data()

circuit_info = session.CircuitInfo

def rotate(xy, *, angle):
    rot_mat = np.array([[np.cos(angle), np.sin(angle)],
                        [-np.sin(angle), np.cos(angle)]])
    return np.matmul(xy, rot_mat)

# Get an array of shape [n, 2] where n is the number of points and the second
# axis is x and y.
track = pos.loc[:, ('X', 'Y')].to_numpy()

# Convert the rotation angle from degrees to radian.
track_angle = circuit_info.rotation / 180 * np.pi

# Rotate and plot the track map.
rotated_track = rotate(track, angle=track_angle)
plt.plot(rotated_track[:, 0], rotated_track[:, 1])

offset_vector = [500, 0]  # offset length is chosen arbitrarily to 'look good'

# Iterate over all corners.
for _, corner in circuit_info.corners.iterrows():
    # Create a string from corner number and letter
    txt = f"{corner['Number']}{corner['Letter']}"

    # Convert the angle from degrees to radian.
    offset_angle = corner['Angle'] / 180 * np.pi

    # Rotate the offset vector so that it points sideways from the track.
    offset_x, offset_y = rotate(offset_vector, angle=offset_angle)

    # Add the offset to the position of the corner
    text_x = corner['X'] + offset_x
    text_y = corner['Y'] + offset_y

    # Rotate the text position equivalently to the rest of the track map
    text_x, text_y = rotate([text_x, text_y], angle=track_angle)

    # Rotate the center of the corner equivalently to the rest of the track map
    track_x, track_y = rotate([corner['X'], corner['Y']], angle=track_angle)

    # Draw a circle next to the track.
    plt.scatter(text_x, text_y, color='grey', s=140)

    # Draw a line from the track to this circle.
    plt.plot([track_x, text_x], [track_y, text_y], color='grey')

    # Finally, print the corner number inside the circle.
    plt.text(text_x, text_y, txt,
             va='center_baseline', ha='center', size='small', color='white')

plt.title(session.event['Location'])
plt.xticks([])
plt.yticks([])
plt.axis('equal')
plt.show()
```
